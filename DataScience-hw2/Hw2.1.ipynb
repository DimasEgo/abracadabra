```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
URL = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
b_rate = pd.read_html(URL, match='Коефіцієнт народжуваності в регіонах України')[0]
b_rate
```

#Замініть у таблиці значення "—" на значення NaN#

```python
b_rate.replace(to_replace='—', value = np.nan, inplace=True)
b_rate
```

#Визначте типи всіх стовпців за допомогою dataframe.dtypes#

```python
b_rate.dtypes
```

#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"#

```python
b_rate[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']] = b_rate[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']].apply(pd.to_numeric)
b_rate.dtypes
```

#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)#

```python
b_rate.isnull().sum()*100/28
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці#
```

```python
b_rate.drop(27)
```

#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)#

```python
b_rate.fillna(b_rate.mean(numeric_only=True)).round(1)
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україн#
```

```python
b_rate[b_rate['2019'] > b_rate['2019'].mean()][['Регіон', '2019'] ]
```

#У якому регіоні була найвища народжуваність у 2014 році?#

```python
b_rate.iloc[b_rate['2014'].idxmax()]['Регіон']
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році#
```

```python
sns.set_style('darkgrid')
plt.figure(figsize = (12,6))
plt.bar(b_rate['Регіон'], b_rate['2019'])
plt.title('Діаграма народжуваності по регіонах у 2019 році')
plt.tick_params(labelrotation = 90)
plt.show()
```
